```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel("moni6.xlsx")
df=pd.DataFrame(data)
print(df)
colors=["red","yellow","orange","violet","green","gray"]
plt.bar(df["Station ID"],df["Station Name"],color=colors)
plt.xlabel("Station ID",fontsize=20)
plt.ylabel("Station  Name",fontsize=20)
plt.title("Delhi Metro Station",fontsize=20)
plt.show()
```
